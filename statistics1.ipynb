## 3. Create a dataset and classify it into nominal, ordinal, interval, and ratio types

| **Data Column**     | **Type of Data** | **Explanation**                                                                    |
| ------------------- | ---------------- | ---------------------------------------------------------------------------------- |
| **Student Name**    | **Nominal**      | Names are just labels, no order or ranking                                         |
| **Gender**          | **Nominal**      | Categories like Male/Female with no numerical or ordered meaning                   |
| **Grade Level**     | **Ordinal**      | Categories have a meaningful order: Excellent > Good > Average > Poor              |
| **Exam Score (°C)** | **Interval**     | Temperature has no true zero, but equal intervals (e.g., 1°C difference is same)   |
| **Marks (%)**       | **Ratio**        | Has a meaningful zero, percentages can be compared and multiplied                  |
| **Height (cm)**     | **Ratio**        | Has a true zero, allows meaningful ratio comparisons (e.g., 160 cm is twice 80 cm) |



```python
import pandas as pd

data = {
    "Student Name": ["Aditi", "Rahul", "Sneha", "Aman", "Priya"],
    "Gender": ["Female", "Male", "Female", "Male", "Female"],
    "Grade Level": ["Excellent", "Good", "Average", "Poor", "Good"],
    "Exam Score (°C)": [32, 28, 30, 29, 31],
    "Marks (%)": [88, 75, 62, 49, 80],
    "Height (cm)": [165, 170, 158, 172, 160]
}

df1 = pd.DataFrame(data)

df1
```

## 23. Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

skewness = df.skew(numeric_only=True)
kurtosis = df.kurt(numeric_only=True)

sns.histplot(skewness, kde=True)
plt.title('Skewness Distribution')
plt.show()

sns.histplot(kurtosis, kde=True)
plt.title('Kurtosis Distribution')
plt.show()
```

## 24. Implement the Pearson and Spearman correlation coefficients for a dataset.

```python
import seaborn as sns
from scipy.stats import pearsonr, spearmanr

df2 = sns.load_dataset('tips')
df2.dtypes
df2.head()

x = df2['total_bill']
y = df2['tip']

pearson_corr= pearsonr(x, y)
spearman_corr= spearmanr(x, y)

print(f"Pearson Correlation Coefficient: {pearson_corr}")

print(f"Spearman Correlation Coefficient: {spearman_corr}")
```
